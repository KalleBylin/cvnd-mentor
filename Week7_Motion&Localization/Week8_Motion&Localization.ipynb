## Robot localization

Big questions:
- **Where am I?**
    - Senses
    - Relates sensor readings to a world model
    - Computes location relative to model
    - Assumes a perfect world model
- **What is the world around me?**
    - Senses from various positions
    - Integrates measurements to produce map
    - Assumes perfect knowledge of position
